## Task 2 - Querying

Use DirectMarketing.xlsx

#### What is the distribution of Customer Salary?

Create a histogram of customer salary for customers that spent more than $500$ dollars.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
amountSpent = pd.read_excel('AmountSpent_13to17.xlsx')
```

```python
# Task 1
average = amountSpent.mean()[1:]
plt.plot(average,label="Average amount spent")
# plt.xticks(np.arange(1,6), ['2013', '2014', '2015', '2016', '2017'])
plt.xlabel("Year")
plt.ylabel("Amount Spent")
plt.title("Line graph of Amount Spent")
```

```python
# Task 2
# What is the distribution of Salary for High Spending Customers?
marketing = pd.read_excel('DirectMarketing.xlsx')
# marketing.describe()
salary = marketing['Salary']
n, bins, patches = plt.hist(salary, rwidth = 0.95, bins=10, align = "mid", label = "Number of Customers", color = [0, 0, 1, 1])
plt.xlabel("Salary")
plt.ylabel("Number of Customers")
plt.title("Histogram of Salary")
plt.legend()
plt.xticks(bins,rotation = 45)
```

```python
# What is the distribution of Salary for High Spending Customers?
# marketing.describe()
greatSpent_plot = plt.figure()
great_spent = marketing.query("AmountSpent > 500")
n, bins, patches = plt.hist(great_spent['Salary'], rwidth = 0.95, bins=10, align = "mid", label = "Number of Customers", color = [0, 0, 1, 1])
plt.xticks(bins,rotation=45)
plt.xlabel("Salary")
plt.ylabel("Number of Customers")
plt.title("Histogram of Salary for customers that \n spent more than  500  dollars")
plt.legend()
```

```python
# Is the amount spent different between male and female customers?
# 1
# marketing.head()
male = marketing.query("Gender == 'Male'")

# male.head()

female = marketing.query("Gender == 'Female'")

n, bins, patches = plt.hist([male['AmountSpent'],female['AmountSpent']],label = ["Male","Female"])
plt.xticks(bins)
plt.legend()
plt.xlabel("Amount Spent")
plt.ylabel("Number of Customers")
plt.title("Histogram of Amount Spent for Different Genders")
```

```python
male_var = male['AmountSpent'].var()
male_average = male['AmountSpent'].mean()
print(male_var)
print(male_average)
```

```python
female_var = female['AmountSpent'].var()
female_average = female['AmountSpent'].mean()
print(female_var)
print(female_average)
```
