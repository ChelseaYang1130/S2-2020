# KNN

**Partial model**

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

NBA_train=NBA_train.drop(['TEAM','POSITION','Train_ID'],1)
NBA_test=NBA_test.drop(['TEAM','POSITION','Test_ID'],1)
```

```python
Y_full = NBA_train['SALARY']
X_full = NBA_train.drop('SALARY',1)
X_trainF, X_valF, Y_trainF, Y_valF = train_test_split(X_full, Y_full, test_size=0.2, random_state=1)
Y_testF = NBA_test['SALARY']
X_testF = NBA_test.drop('SALARY',1)
```

```python
values = np.arange(1, 101)
knnF_valerrors = []
for k in values:

    knn = KNeighborsRegressor(n_neighbors= k) 
    knn.fit(X_trainF, Y_trainF)

    knnF_pred = knn.predict(X_valF)

    val_error = np.sqrt(mean_squared_error(Y_valF, knnF_pred))

    knnF_valerrors.append(val_error)
    
fig, ax= plt.subplots()
ax.plot(values, knnF_valerrors)
ax.set_xlabel('Number of neighbours')
ax.set_ylabel('Test error')
plt.show()
```

```python
# 𝑘  with lowest validation error
k = 1 + np.argmin(knnF_valerrors)
knn9F = KNeighborsRegressor(n_neighbors= k) 
knn9F.fit(X_trainF, Y_trainF)
pred_knn9F = knn9F.predict(X_testF)
test_errorF = np.sqrt(mean_squared_error(Y_testF, pred_knn9F))
test_errorF
```

# Linear regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
def linearVal(var):
    Y = NBA_train['SALARY']
    X = NBA_train.drop('SALARY',1)[var]
    X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=1)
    Y_test = NBA_test['SALARY']
    X_test = NBA_test.drop('SALARY',1)[var]

    max_deg = 10

    validation_error = {}

    for i in range(1, max_deg):

        poly_transformer = PolynomialFeatures(i) 
        poly_x = poly_transformer.fit_transform(X_train)

       
        lin_reg = LinearRegression()

        # Estimate coefficients
        lin_reg.fit(poly_x, Y_train)
        
        Y_predTrain = lin_reg.predict(poly_x)
        poly_xVal = poly_transformer.fit_transform(X_val)
        # Calculate predictions
        Y_predVal = lin_reg.predict(poly_xVal)

        mse_valid = np.sqrt(mean_squared_error(Y_val, Y_predVal))
        validation_error[i]=mse_valid

    min_error = np.inf

    for i in validation_error:
        if validation_error[i]<min_error:
            min_error = validation_error[i]
            degree = i
    
    poly_transformer = PolynomialFeatures(degree) 
    poly_x = poly_transformer.fit_transform(X_train)

    lin_reg = LinearRegression()

    lin_reg.fit(poly_x, Y_train)        
    poly_xTest = poly_transformer.fit_transform(X_test)
        
    Y_predTest = lin_reg.predict(poly_xTest)

    mse_test = np.sqrt(mean_squared_error(Y_test, Y_predTest))
    
    return degree,min_error,mse_test
```

```python
full_model = list(NBA_train.drop('SALARY',1).columns)
print(linearVal(full_model))
```

```python
lin_error = []
# pd.DataFrame(columns=['Variable','Validation error','Polynomial degree'])
for i in full_model:
    degree,val_error,test_error = linearVal([i])
    lin_error.append([i,degree,val_error,test_error])

pd.DataFrame(lin_error,columns=['Variable','Polynomial degree','Validation error','Test error']).sort_values(by=['Validation error'])
```
